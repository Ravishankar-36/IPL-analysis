```python
MI=['R Sharma','S Rutherford','SA Yadav','CA Lynn','SS Tiwary','DS Kulkarni','JJ Bumrah','SL Malinga','MJ McClenaghan','RD Chahar',
   'TA Boult','HH Pandya','J Yadav','AP Tare','NM Coulter-Nile','Ishan Kishan','Q de Kock','KH Pandya','KA Pollard']

MI_data_missing=['Anmolpreet singh','Mohsin khan','Balwant rai singh','Digvijay Deshmukh','Ankul Roy']
```

```python
CSK=['MS Dhoni','RA Jadeja','M Vijay','KM Jadhav','KV Sharma','PP Chawla','AT Rayudu','SK Raina','Imran Tahir','DL Chahar',
    'F du Plessis','SN Thakur','M Santner','DJ Bravo','S Curran','SR Watson','Harbhajan Singh']
CSK_data_missing=['Narayan jagadeesan','Ruturaj gaikwad','KM Asif','Josh hazlewood','Lungi Ngidi','Monu kumar','Sai kishore']
```

```python
cons_df=dfipl1[['batsman','bowler','batsman_runs']]
```

```python
cons_df['ball_count']=dfipl2['batsman_runs']
```

```python
cons_df.columns=['batsman','bowler','avgrunsperball','totalballsfaced']
```

```python
cons_df['combi']=cons_df['batsman']+cons_df['bowler']
```

```python
cons_df
```

```python
from itertools import product

a=list(product(MI, CSK))
b=list(product(CSK, MI))
```

```python
MI_CSK_combi=pd.DataFrame(a)
MI_CSK_combi['team']='MI'
CSK_MI_combi=pd.DataFrame(b)
CSK_MI_combi['team']='CSK'
frame=[CSK_MI_combi,MI_CSK_combi]
comb=pd.concat(frame)
```

```python
comb.columns=['p1','p2','p1_team']
comb['playcomb']=MI_CSK_comb['p1']+MI_CSK_comb['p2']
```

```python
players_combi=comb['playcomb'].to_list()
```

```python
final=cons_df[cons_df['combi'].isin(players_combi)]
```

```python
players=final['batsman'].to_list()
teamtag=[]
for a in players:
    if a in MI:
        teamtag.append('MI')
    else:
        teamtag.append('CSK')

final['team']=teamtag        
```

```python
final
```

```python
import numpy as np
weighted_score=final.groupby(["batsman","team"]).apply(lambda x: np.average(x['avgrunsperball'], weights=x['totalballsfaced']))
```

```python
ws=pd.DataFrame(weighted_score)
```

```python
ws.reset_index()
```

```python
ws.to_csv('weighted_score.csv')
```
